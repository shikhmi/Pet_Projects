```python
import requests
from parsel import Selector
```

```python
url = "https://ru.wikipedia.org/wiki/Категория:Фильмы_по_алфавиту"
html = requests.get(url, headers={"User-Agent": "Mozilla/5.0"}).text
sel = Selector(text=html)
```

```python
pager_links = sel.css("#mw-pages a::text").getall()
print("pager texts sample:", [t.strip() for t in pager_links if t.strip()][:20])
```

```python
next_page = sel.xpath("//div[@id='mw-pages']//a[normalize-space()='Следующая страница']/@href").get()
print("next page:", next_page)
```
